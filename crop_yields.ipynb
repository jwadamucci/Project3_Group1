```python
import pandas as pd
```

```python
# Load the two CSV files into DataFrames
yield_df_path = '/Users/GURU/Desktop/Project 3/Resources/yield_df.csv'  # Replace with actual path
yield_path = '/Users/GURU/Desktop/Project 3/Resources/yield.csv'       # Replace with actual path

yield_df = pd.read_csv(yield_df_path)
yield_data = pd.read_csv(yield_path)
```

```python
# Display the columns for both datasets
print("Columns in yield_df.csv:", yield_df.columns.tolist())
print("Columns in yield.csv:", yield_data.columns.tolist())
```

```python
# Concatenate the two DataFrames
merged_yield = pd.concat([yield_df, yield_data], ignore_index=True)
print("Successfully merged yield_df.csv and yield.csv.")
```

```python
# Check for missing data
print("Missing values in the merged DataFrame:")
print(merged_yield.isnull().sum())
```

```python
# Optionally, inspect the first few rows of the merged DataFrame
print("First few rows of the merged DataFrame:")
print(merged_yield.head())
merged_yield.head()
```

```python
# Save the merged DataFrame into crop_yields.csv
output_file = '/Users/GURU/Desktop/Project 3/Resources/crop_yields.csv'  # Replace with your desired output path
merged_yield.to_csv(output_file, index=False)

print(f"Merged dataset saved as {output_file}")
```

```python
# Shape of the DataFrame
print("Shape of the merged DataFrame:", merged_yield.shape)

# Data types of the columns
print("\nData types of the merged DataFrame:")
print(merged_yield.dtypes)

# Summary statistics
print("\nSummary statistics of the merged DataFrame:")
print(merged_yield.describe(include='all'))

# Unique values in specific columns (adjust if needed)
for column in ['crop', 'year', 'country', 'yield', 'unit', 'area harvested']:
    if column in merged_yield.columns:
        print(f"\nUnique values in the '{column}' column:")
        print(merged_yield[column].unique())
```
