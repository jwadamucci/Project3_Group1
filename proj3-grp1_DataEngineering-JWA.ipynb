```python
# Dependencies
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
```

```python
# load files into pandas
yield_data = pd.read_csv("Resources/yield.csv")
yield_df_data = pd.read_csv("Resources/yield_df.csv")
pesticides_data = pd.read_csv("Resources/pesticides.csv")
rainfall_data = pd.read_csv("Resources/rainfall.csv")
temp_data = pd.read_csv("Resources/temp.csv")
```

```python
#create and fiew dataframe
yield_df = pd.DataFrame(yield_data)
yield_df
```

```python
#create and fiew dataframe
yield_df_df = pd.DataFrame(yield_df_data)
yield_df_df
```

```python
#create and fiew dataframe
pesticides_df = pd.DataFrame(pesticides_data)
pesticides_df
```

```python
#create and fiew dataframe
rainfall_df = pd.DataFrame(rainfall_data)
rainfall_df = rainfall_df.rename(columns={' Area':'Area'})
rainfall_df
```

```python
#create and fiew dataframe
temp_df = pd.DataFrame(temp_data)

#rename country column to Area in the temp_df
temp_df = temp_df.rename(columns={'country':'Area'})
temp_df = temp_df.rename(columns={'year':'Year'})
temp_df
```

```python
#merge pesticides/rainfall/temp dataframes into one environment dataframe
print(pesticides_df.columns.tolist())
print(rainfall_df.columns.tolist())
print(temp_df.columns.tolist())

interim_df = pd.merge(pesticides_df,rainfall_df, on=['Area','Year'])
interim_df
# enviroment_df = pd.merge(interim_df, temp_df, on=['Area','Year'])
# enviroment_df
```

```python
enviroment_df = pd.merge(interim_df, temp_df, on=['Area','Year'])
enviroment_df
```

```python
#check for blanks (totals should be zero!)
print(enviroment_df.isnull().sum())
```

```python
# merge yield data with the environmentals
final_data_df = pd.merge(yield_df, enviroment_df, on=['Year', 'Area'])
final_data_df.head(2)
```

```python
# clean up the final dataframe.
print(final_data_df.columns)

final_data_df = final_data_df.drop(columns=['Area Code','Domain Code', 'Domain_x','Element Code','Item Code','Year Code','Domain_y','Element_y', 'Item_y', 'Unit_y','Element_x', 'Unit_x'])

print(final_data_df.columns)
```

```python
final_data_df.head(2)
final_data_df = final_data_df.rename(columns={'Item_x':'Crop','Value_x':'Crop_Yield','Value_y':'Pesticide_hg/ha','average_rain_fall_mm_per_year':'Rain_mm','avg_temp':'Temp_c'})
final_data_df
```

```python
final_data_df['Temp_c'] = final_data_df['Temp_c'].astype(float)
final_data_df['Rain_mm'] = pd.to_numeric(final_data_df['Rain_mm'], errors='coerce')

print(final_data_df.info())

final_data_df.to_csv('Resources/final_crop_data.csv')

final_data_df
```
